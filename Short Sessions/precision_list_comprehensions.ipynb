````python

from math import log


def flatten_and_square(matrix):
    """
    Given a matrix (list of lists), return a single list
    containing the square of every number, 
    but ignore non-positive numbers.
    """

    result = [x**2 for row in matrix for x in row if x > 0]

    print(result)

    # now doing the traditional way without any list comprehensions

    results = []

    for row in matrix:
        for x in row:
            if x > 0:
                results.append(x**2)
    return result, results

if __name__ == "__main__":
    matrix = [
        [1, -2, 3],
        [-4, 5, 6],
        [0, -7, 8]
    ]

    flattened_list, traditional_list = flatten_and_square(matrix)
    print("Using list comprehension:", flattened_list)
    print("Using traditional method:", traditional_list)
````
