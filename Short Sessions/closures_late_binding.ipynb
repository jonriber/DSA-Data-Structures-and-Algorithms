Todays topica is this: Closures & late binding

Tricky topic where bugs might appear misteriously.

## Step 1

The following code is this:

```python

    funcs = []

    for i in range(3):
        def f():
            return i
        funcs.append(f)
    
    print([fn() for fn in funcs])
```

After running this code, the expected result wold naturally be : [0,1,2], but instead, it is [2,2,2]

My task is to fix the code to function as expected.

## Step 2 - Constraints

- Must keep a loop
- must keep functions created inside the loop
- May not use global variables
- solve it in two different ways

## Step 3 - Learning

- What closure actually captures
- difference between name binding and value binding
- why python resolves variables at call time, not definition time
- default arguments and function factoreis hange binding behavior

````python
# the code, this is the default solution

funcs = []

for i in range(3):
    def f(i=i):
        return i
    funcs.apppend(f)

print([fn() for fn in funcs])
````

````python
# Second solution - more explict

def make_f(value):
    def f():
        return value
    return f

funcs = []

for i in range(3):
    funcs.append(make_f(i))

print([fn() for fn in funcs])
````
