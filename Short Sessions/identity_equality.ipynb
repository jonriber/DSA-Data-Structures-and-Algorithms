### Rule

## Integers: the weirdness

Case A

x == y --> true always (same value)

x is y --> is often true in cpython because cpython interns reuses small integers for efficiency, so both names may point to the same cached integer object


Case B

x == y --> still true

x is y --> is often false because 257 typically isnt in that small int cache, so two separate integer objects may be created

`important`: is behavior for numbers/strings can vary with implementation details. Thats exactly why using is for value comparison is a footgun.

## Step 4 - Task answers

- Difference between is and == 
    - == asks: "Do these two things have the same value/content?"
    - is asks: "Are these literally the same object?"

- Why integers behave differently
    - Because Cpython may reuse (cache/intern) certain objects like small integers and some strings. That makes comparisons look consistent sometimes, but its not a guarantee you should rely on for logic.

- One rule for when to use is
    - use is only for:
        - checking against singletons like None, True, Flase
        - Checking if two variables refer to the same object on purpose
        - sentinel objects you created
