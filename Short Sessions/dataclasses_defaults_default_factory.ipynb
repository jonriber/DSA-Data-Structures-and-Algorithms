## Step 1: code interpretation

```python
from dataclasses import dataclass

```

````python
from dataclasses import dataclass

@dataclass
class User:
    name: str
    tags: list = []

u1 = User("Alice")
u2 = User("Bob")

u1.tags.append("admin")

print(u1.tags)
print(u2.tags)
````

## Step 2: The task

- Explain exactly what is happening
- Fix the dataclass correctly
- Each user gets their own list
- No shared mutable state
- dont initialize the list in __init__

## Step 3: The constraint

- Must use `default_factory`
